```python
import DB as db
import ipywidgets as widgets
from IPython.display import display


query_add_url = '''
insert into fgroups (
lang,
sug_kvutza,
url,
branch,
group_title,
description,
ezor,
status_kvuza
)
values(
{lang},
{url},
{branch},
{group_title},
{description},
{ezor},
{status_kvuza}
)
'''

def add_url(lang, sug_kvutza, url, branch, group_title, description, ezor, status_kvuza): 
    
    query = str(f''' 
        insert into fbgroups (
        lang,
        sug_kvutza,
        url,
        branch,
        group_title,
        description,
        ezor
        )
        values(
        '{lang}',
        '{sug_kvutza}',
        '{url}',
        '{branch}',
        '{group_title}',
        '{description}',
        '{ezor}'
        )''')
    
    db.execute(query, do='insert')

def get_dublicates():
    query = 'SELECT * FROM `fbgroups` WHERE `url` IN (SELECT `url` FROM `fbgroups` GROUP BY `url` HAVING COUNT(*) > 1) ORDER BY `url`'
    df = db.execute(query , do='select')
    print(df)
 
def delete_ROWID(ROWID):
    query = str(f'DELETE FROM `fbgroups` where ROWID =  {ROWID}')
    print(query)
    db.execute(query , do='delete')
    get_dublicates()
    
def render_inputs():
    '''
    виджеты 
    '''
    lang = widgets.RadioButtons(
    options=['he', 'ru'],
    #    value='pineapple', # Defaults to 'pineapple'
    #    layout={'width': 'max-content'}, # If the items' names are long
    description='Lang',
    disabled=False
    )

    display(lang)

    ''' ------------------------------------------------------ '''
    sug_kvutza = widgets.RadioButtons(
    options=['pirsum', 'mexirot'],
    #    value='pineapple', # Defaults to 'pineapple'
    #    layout={'width': 'max-content'}, # If the items' names are long
    description='sug_kvutza',
    disabled=False
    )

    display(sug_kvutza)
    ''' ------------------------------------------------------ '''
    url = widgets.Text(
        value='',
        placeholder='',
        description='URL' ,
        disabled=False   
    )
    display(url)
    ''' ------------------------------------------------------ '''
    branch = widgets.RadioButtons(
    options=['aluf', 'tzama'],
    #    value='pineapple', # Defaults to 'pineapple'
    #    layout={'width': 'max-content'}, # If the items' names are long
    description='sug_kvutza',
    disabled=False
    )
    display(branch)
    ''' ------------------------------------------------------ '''
    group_title = widgets.Text(
        value='',
        placeholder='',
        description='group_title' ,
        disabled=False   
    )
    display(group_title)
    ''' ------------------------------------------------------ '''
    description = widgets.Text(
        value='',
        placeholder='',
        description='description' ,
        disabled=False   
    )
    display(description)
    ''' ------------------------------------------------------ '''
    ezor = widgets.RadioButtons(
        options=['Рамла', 'Ришон лецион' , 'Реховот', 'Несциона','Явне','כפר-סבא>הוד-השרון>רעננה' , 'רעננה, הרצליה, רמת השרון',
                 'Тель Авив','Рамат Ган','Центр' , 'Израиль' , 'Нетания','Юг', ' Тель авив',
                'Холон','Петах Тиква'],
        #    value='pineapple', # Defaults to 'pineapple'
        #    layout={'width': 'max-content'}, # If the items' names are long
        description='',
        disabled=False
    )
    display(ezor)
    ''' ------------------------------------------------------ '''
    status_kvuza = widgets.RadioButtons(
        options=['active'],
        #    value='pineapple', # Defaults to 'pineapple'
        #    layout={'width': 'max-content'}, # If the items' names are long
        description='',
        disabled=False
    )
    display(status_kvuza)
    ''' ------------------------------------------------------ '''


    def on_button_save_clicked(b): # Описываем обработчик события
        add_url(lang.value, url.value, sug_kvutza.value, branch.value, group_title.value, description.value, ezor.value, status_kvuza.value)
        url.value = ''
        get_dublicates()

    button_save = widgets.Button(description="Save")
    button_save.on_click(on_button_save_clicked) # Назначаем этот обработчик на событие "on_click"
    display(button_save)

render_inputs()

get_dublicates()

textbox_ROWID = widgets.Text(
        value='',
        placeholder='',
        description='ROWID' ,
        disabled=False   
    )
display(textbox_ROWID)
    
def button_delete_ROWID_clicked(b): # Описываем обработчик события
    delete_ROWID(textbox_ROWID.value)
    get_dublicates()

button_delete_ROWID = widgets.Button(description="DELETE")
button_delete_ROWID.on_click(button_delete_ROWID_clicked) # Назначаем этот обработчик на событие "on_click"
display(button_delete_ROWID)

```
